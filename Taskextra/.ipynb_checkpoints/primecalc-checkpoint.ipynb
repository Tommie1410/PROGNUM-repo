```python
#prime calcuator
from math import *
list_primes = []

def isprime(num):
    if (num <= 1):
        return(False)
    else:
        for i in range(2, int(sqrt(num)) + 1):
            if (num % i == 0):
                return(False)
            else:
                return(True)
        return(True)

num, count = 0, 0
while count < 60:
    if (isprime(num)):
        list_primes.append(num)
        count +=1
        num +=1
    else:
        num +=1

print(list_primes)
```

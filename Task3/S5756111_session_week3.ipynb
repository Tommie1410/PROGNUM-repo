```python
from math import *
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
m_v_dict = {}

print("The esape velocity for solar system objects: \n")
for i in range(len(masses)):
    v_esc = sqrt((2*G*masses[i])/(1000*radii[i]))
    print(f"{names[i]}: {v_esc:.2f} m/s") 
    print(f"mass: {masses[i]:.3} \n")
    m_v_dict[names[i]] = v_esc
print(m_v_dict)
```

```python
#Convergence of sequences
def inverse_sq_sum(N):
    s = 0
    for i in range(1,N):
        x = 1/(i**2)
        s = s + x
    return(s)

list_N = []

print('Input the values for which the inverse square sum will be calculated, finish with 0 (zero)')

while (True):
    a = int(input())
    if (a == 0):
        break
    else:
        list_N.append(a)

a = 0

for i in range(len(list_N)):
    print(inverse_sq_sum(list_N[i]))
    b = inverse_sq_sum(list_N[i])
    print(f'Diference from last term: {b - a}')
    a = b
    print()
    
print('If the difference is getting smaller, the sum likely converges\nEnd of program')
    
    
```

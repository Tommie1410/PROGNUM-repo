```python
#Convergence of sequences
def inverse_sq_sum(N):
    s = 0
    for i in range(1,N):
        x = 1/(i**2)
        s = s + x
    return(s)

list_N = []

print('Input the values for which the inverse square sum will be calculated in ascending order')
print('confirm with Enter, finish with 0 (zero)')

order, b = 0, 0

while (True):
    N = int(input())
    if (N < b):
        print('Input the numbers in ascending order!')
        order = 1
        break
    else:
        if (N == 0):
            break
        else:
            list_N.append(N)
            b = N

a = 0

if (order == 0):
    for i in range(len(list_N)):
        print(inverse_sq_sum(list_N[i]))
        b = inverse_sq_sum(list_N[i])
        print(f'Diference from last term: {b - a}')
        a = b
        print()

    diff1 = inverse_sq_sum(list_N[-2]) - inverse_sq_sum(list_N[-1])
    diff2 = inverse_sq_sum(list_N[-1]) - inverse_sq_sum(list_N[len(list_N)-1])

    if (diff1 > diff2):
        print('The sum diverges')
    else:
        print('The sum converges')
    
print('\nEnd of program')
    
    
```

```python
#Area and volume of a cylinder
from scipy.constants import pi

def Area(r, h):
    A = 2*pi*r**2 + 2*pi*r*h
    return(A)

def Volume(r, h):
    V = pi*(r**2)*h
    return(V)

def cylinder(r, h):
    print (f'The area fo the cylinder is: {Area(r, h)}')
    print(f'Its volume is: {Volume(r, h)}')
    print('\nEnd')

r = float(input('Input the radius of the cylinder base: '))
h = float(input('Input the height of the cylinder : '))
cylinder(r, h)
```

```python
#Std dev using ndarray and in a one-liner
import numpy as np

def stdev(x):
    return(np.sqrt((1/1000)*sum((x - sum(x)/1000)**2)))

x = np.random.normal(1, 5, 1000)
print('stdev: ', stdev(x))
print('numpy std: ', np.std(x))
```

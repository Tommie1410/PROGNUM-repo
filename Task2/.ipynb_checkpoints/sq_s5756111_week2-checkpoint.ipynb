```python
#Question 15
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

set_night1 = set(night1)
set_night2 = set(night2)

inters = set.intersection(set_night1, set_night2)
list_inters = list(inters)

print(list_inters)

if (5 in list_inters):
    print ('5 is in the intesection list')
else:
    print ('5 is not in the intersection list')

if (5 in set_night1):
    print('5 is in the set version of the list night1')
else:
    print('5 is not in the set version of the list night1')
```

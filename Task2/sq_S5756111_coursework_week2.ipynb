```python
#Julian date
print('This is the Julian date calcualtor')
print('Enter the start year')
Ys = int(input())
print('Enter the start month')
Ms = int(input())
print("Enter the start day")
Ds = float(input())

print('Enter the end year')
Ye = int(input())
print('Enter the end month')
Me = int(input())
print("Enter the end day")
De = float(input())

JDs = 367*Ys-7*((Ys+(Ms+9)/12)/4)-3*((Ys+(Ms-9)/7)/100+1)/4+(275*Ms)/9+Ds+1721029-0.5

JDe = 367*Ye-7*((Ye+(Me+9)/12)/4)-3*((Ye+(Me-9)/7)/100+1)/4+(275*Me)/9+De+1721029-0.5

print('The Julian date difference is:', JDe - JDs)
```

```python
# Roots of second degree polynomials
from math import *

print('This tools solves roots of polynomials of the form ax^2 + bx +c')
print('Enter coefficients a, b and c respectively')
a = float(input())
b = float(input())
c = float(input())

D = b**2 - 4*a*c

if (D > 0):
    x1 = (-b + sqrt(D))/(2 * a)
    x2 = (-b - sqrt(D))/(2 * a)
    print (f'The solutions are: {x1}, {x2}')
elif (D == 0):
    x1 = -b/(2 * a)
    print(f'The solution is: {x1}')
else:
    print(f'There is no solution within the real numbers')
```

```python
#Question 2.16

night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]

set_night1 = set(night1)
set_night2 = set(night2)

inters = set.intersection(set_night1, set_night2)
list_inters = list(inters)

diff = set.difference(set_night1, set_night2)
list_diff = list(diff)

both = set(night1 + night2)
list_both = list(both)

night1_excl = set(night1) - diff
list_night1_excl = list(night1_excl)

print('List intersection:', list_inters)
print('List difference:', list_diff)
print('List of both sets:', list_both)
print('List of night1 without elements not availible in night 2:', list_night1_excl)
```

## Q 1.8: Magnitude calculations

```python
# Sirius data
# apparentMagnitude = -1.46
# absoluteMagnitude = 1.45

print ('input apparent magnitude: ')
apparentMagnitude = float(input())

print('input absolute magnitude: ')
absoluteMagnitude = float(input())

# The distance is related to the magnitudes as m-M=5.Log(d/10)
# 1 Parsec = 3.26164 ly

m = apparentMagnitude
M = absoluteMagnitude

d = 10.0 * pow( 10.0, (m-M)/5.0 ) * 3.26164

print (d)
```
